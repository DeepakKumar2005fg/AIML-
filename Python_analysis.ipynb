```python
#Practical 2: Handle Missing Values
df = pd.DataFrame({
    "Name": ["A", "B", "C", "D"],
    "Age": [20, None, 22, 21],
    "Marks": [85, 90, None, 88]
})

print("Before:\n", df)

# Fill missing values
df["Age"].fillna(20, inplace=True)
df["Marks"].fillna(40, inplace=True)

print("After:\n", df)
```

```python
# Basic Data Analysis Program

import pandas as pd
import matplotlib.pyplot as plt

# Step 1: Load data (you can also read from CSV, Excel, etc.)
data = {
    'Name': ['Amit', 'Neha', 'Rahul', 'Priya', 'Rohit'],
    'Age': [23, 25, 22, 28, 24],
    'Score': [85, 90, 78, 88, 95]
}

df = pd.DataFrame(data)

# Step 2: Display dataset
print("Dataset:")
print(df)

# Step 4: Sorting by Score
print("\nSorted by Score:")
print(df.sort_values(by='Score', ascending=False))

# Step 5: Visualization
plt.bar(df['Name'], df['Score'], color='skyblue')
plt.xlabel('Name')
plt.ylabel('Score')
plt.title('Scores of Students')
plt.show()
```
