```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis as LDA
from sklearn.decomposition import FactorAnalysis, FastICA

# 1. Load the Iris dataset and apply LDA
iris = load_iris()
X_iris = iris.data
y_iris = iris.target

# Apply LDA
lda = LDA(n_components=2)
X_lda = lda.fit_transform(X_iris, y_iris)

# Convert to DataFrame for visualization
lda_df = pd.DataFrame(data=X_lda, columns=['LD1', 'LD2'])
lda_df['target'] = y_iris

# Plot LDA results
plt.figure(figsize=(8, 5))
colors = ['red', 'green', 'blue']
for i in np.unique(y_iris):
    plt.scatter(lda_df[lda_df['target'] == i]['LD1'], lda_df[lda_df['target'] == i]['LD2'], label=iris.target_names[i], color=colors[i])
plt.title('Linear Discriminant Analysis (LDA)')
plt.xlabel('LD1')
plt.ylabel('LD2')
plt.legend()
plt.grid()
plt.show()

# 2. Generate random data and apply Factor Analysis
np.random.seed(0)
X_random = np.random.rand(100, 5)

# Apply Factor Analysis
fa = FactorAnalysis(n_components=2)
X_fa = fa.fit_transform(X_random)

# Convert to DataFrame for analysis
fa_df = pd.DataFrame(data=X_fa, columns=['Factor1', 'Factor2'])

# Plot Factor Analysis results
plt.figure(figsize=(8, 5))
plt.scatter(fa_df['Factor1'], fa_df['Factor2'], alpha=0.7)
plt.title('Factor Analysis')
plt.xlabel('Factor1')
plt.ylabel('Factor2')
plt.grid()
plt.show()

# 3. Generate mixed signals and apply ICA
S = np.array([[1, 1, 1], [1, 2, 1], [2, 1, 1]])  # Source signals
A = np.array([[1, 1, 0], [0, 1, 1], [1, 0, 1]])  # Mixing matrix
X_mixed = S.dot(A.T)  # Mixed signals

# Apply ICA
ica = FastICA(n_components=3)
S_ = ica.fit_transform(X_mixed)  # Recovered signals

# Convert to DataFrame for analysis
ica_df = pd.DataFrame(data=S_, columns=['Source1', 'Source2', 'Source3'])

# Plot ICA results
plt.figure(figsize=(8, 5))
plt.subplot(3, 1, 1)
plt.title('Independent Component Analysis (ICA) - Source Signals')
plt.plot(S[0], label='Source 1')
plt.plot(S[1], label='Source 2')
plt.plot(S[2], label='Source 3')
plt.legend()

plt.subplot(3, 1, 2)
plt.title('Mixed Signals')
plt.plot(X_mixed[:, 0], label='Mixed Signal 1')
plt.plot(X_mixed[:, 1], label='Mixed Signal 2')
plt.plot(X_mixed[:, 2], label='Mixed Signal 3')
plt.legend()

plt.subplot(3, 1, 3)
plt.title('Recovered Signals from ICA')
plt.plot(S_[:, 0], label='Recovered Source 1')
plt.plot(S_[:, 1], label='Recovered Source 2')
plt.plot(S_[:, 2], label='Recovered Source 3')
plt.legend()

plt.tight_layout()
plt.show()
```
